<pre><b>
When we search or query something, we have to pass the user agent(data about the user) in the header. Without this, Google will not allow use to access. Based on the user agent value only, google finds out if we login from different device.

We can use parse to pass the values outside of url. If we want to send the data with url, we don't need parse at all
</b></pre>

```python
import urllib.parse as par
```

```python
# Method not allowed. Because Google will not allow the program to fetch the data
url = "https://www.google.com/search"
values = {"q":"Barcelona"}
data = par.urlencode(values)
data = data.encode('utf-8') # Converting data into bytes
req = ure.Request(url,data)
res = ure.urlopen(req)
response_view = res.read()
print(response_view)
```

<pre><b>
When you send a data to a server, we have to convert the data into bytes. We have to convert the string into bytes in particular format like utf-8.
Each character is made of 1 byte or 8 bits. 
UTF encoding - each character is converted into ASCII code and then it will get appended into browser.
Data travels from one machine(our computer) to another machine(browser) in the form of bits/bytes/packets.

<b><pre>Use try catch when you access any server.
We have to pass the user agent when you query something from a website
Default user agent will go as python
</pre></b>

```python
import urllib
```

```python
# Google, user agent value for chrome
# https://developer.chrome.com/multidevice/user-agent
try:
    url = "https://www.google.com/search?q=Barca"
    headers = {}
    headers['user-agent'] = "Mozilla/5.0 (Linux; <Android Version>; <Build Tag etc.>) AppleWebKit/<WebKit Rev> (KHTML, like Gecko) Chrome/<Chrome Rev> Mobile Safari/<WebKit Rev>"
    req = urllib.request.Request(url, headers = headers)
    res = urllib.request.urlopen(req)
    response_data = res.read()
    print(response_data)
except Exception as e:
    print(str(e))
```

```python
try:
    f = open("output-from-google.txt","w")
    f.write(str(response_data))
    f.close()
except Exception as e:
    print(str(e))
```
