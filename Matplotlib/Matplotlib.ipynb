## Object Oriented

```python
# It is like an imaginary canvas. We can add axis to it.
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #Left, Bottom, Width and Height of Axes
#10% from left, 10% from bottom, 80% canvas size from width and height. All these values should be 0 to 1
axes.plot(x,y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("New Title")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.5,0.2,0.3,0.2])
axes1.plot(x,y)
axes1.set_title("Larger plot")
axes2.plot(y,x)
axes2.set_title("Smaller plot")
```

## Sub plots

```python
fig,axes = plt.subplots()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=3)
plt.tight_layout() # It takes care of overlapping layout
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
```

```python
axes # it is an array. We can iterate through it
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title("First plot")
axes[1].plot(y,x)
axes[1].set_title("Second plot")
```

## Figure size and DPI

```python
fig = plt.figure(figsize=(5,2), dpi=100)
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(figsize=(5,2))
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1, figsize=(5,2))
axes[0].plot(x,y)
axes[1].plot(x,y)
plt.tight_layout()
```

## Legends
#### It will show you what plot is a what plot

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label="X Squared")
axes.plot(x,x**3,label="X Cubed")

axes.legend()
```

Notice how are legend overlaps some of the actual plot!

The **legend** function takes an optional keyword argument **loc** that can be used to specify where in the figure the legend is to be drawn. The allowed values of **loc** are numerical codes for the various places the legend can be drawn. See the [documentation page](http://matplotlib.org/users/legend_guide.html#legend-location) for details. Some of the most common [**loc**](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.legend.html) values are:

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label="X Squared")
axes.plot(x,x**3,label="X Cubed")

axes.legend(loc=10)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label="X Squared")
axes.plot(x,x**3,label="X Cubed")

axes.legend(loc=(0.1,0.1))
```
