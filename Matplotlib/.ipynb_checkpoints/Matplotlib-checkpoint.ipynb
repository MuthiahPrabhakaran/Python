```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

### We can create plot in function oriented method and object oriented method
### object oriented is a better way

# Functional method:

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.show() # It is show this in a normal program, but not needed in Jypyter. show is like print method. No "Out" on the cell 
```

```python
plt.plot(x,y)
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.title("My First Plot")
plt.show()
```

```python
plt.subplot(1,2,1) #1 row by 2 column, plot number is the 3rd argument
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(x,y)
```

## Object Oriented

```python
# It is like an imaginary canvas. We can add axis to it.
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #Left, Bottom, Width and Height of Axes
#10% from left, 10% from bottom, 80% canvas size from width and height. All these values should be 0 to 1
axes.plot(x,y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("New Title")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.5,0.2,0.3,0.2])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.5,0.2,0.3,0.2])
axes1.plot(x,y)
axes1.set_title("Larger plot")
axes2.plot(y,x)
axes2.set_title("Smaller plot")
```
